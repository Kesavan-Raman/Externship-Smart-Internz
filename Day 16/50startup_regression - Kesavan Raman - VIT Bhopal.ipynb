```python
#Splitting the X and Y for Prediction Purpose

x = df.iloc[:, 0:4].values
```

```python
y = df.iloc[:, 4].values
```

```python
#Do the Column Trnasfor for Handling the Categorical Values
```

```python
#importing the modules for column transoform

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([("oh", OneHotEncoder(),[3])], remainder = "passthrough")
```

```python
x = ct.fit_transform(x)
```

```python
x
```

```python
y = y.reshape(-1, 1)
```

```python
y
```

```python
#Splitting the test and train Dataset

from sklearn.model_selection import train_test_split
```

```python
x.shape
```

```python
y.shape
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state = 0)
```

```python
#Model Building

from tensorflow.keras.models import Sequential   #used to initialize the model
from tensorflow.keras.layers import Dense, Dropout        #adding layers
```

```python
model = Sequential()
```

```python
import numpy as np

x_train=np.asarray(x_train).astype(np.float32) 
y_train=np.asarray(y_train).astype(np.float32)
x_test=np.asarray(x_test).astype(np.float32) 
y_test=np.asarray(y_test).astype(np.float32)
```

```python
model.add(Dense(units = 6, kernel_initializer = "normal", activation = "relu"))
model.add(Dense(units = 12, kernel_initializer = "normal", activation = "relu"))
model.add(Dense(units = 6, kernel_initializer = "normal", activation = "relu"))
model.add(Dense(units = 3, kernel_initializer = "normal", activation = "relu"))

model.add(Dense(units = 1, kernel_initializer = "normal", activation = "linear"))
```

```python
model.compile(optimizer = "rmsprop", loss = "mse", metrics = ['mae', 'mse'])
```

```python
history = model.fit(x_train, y_train, batch_size = 32, epochs = 1000, validation_split = 0.2)
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['mse'], label = "train_Accuracy")
plt.plot(history.history['val_loss'], label = 'test_accuracy')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import explained_variance_score
```

```python
# accuracy = accuracy_score(y_pred.round(), y_test)

accuracy = explained_variance_score(y_test, y_pred)
```
