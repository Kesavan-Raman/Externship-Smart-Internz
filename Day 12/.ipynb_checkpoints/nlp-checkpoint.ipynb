```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Restaurant_Reviews.tsv", delimiter = "\t")
```

```python
df
```

```python
#import the NLP Library

import re # used to replace the special characters
import nltk # natural language toolkit - all the necessary library for nlp operations

nltk.download("stopwords") #stores in nltk corpus
from nltk.corpus import stopwords #used to detect the stopwords
from nltk.stem.porter import PorterStemmer # Used to Stem our words

ps = PorterStemmer()
```

```python
ps
```

```python
data = []
```

```python
#Preprocessing the Data
```

```python
for i in range(0, 1000):
    review = df["Review"][i]
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words("english"))]
    review = " ".join(review)
    data.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 2000)
```

```python
x = cv.fit_transform(data).toarray()
```

```python
x
```

```python
x.shape
```

```python
y = df.iloc[:, 1:2].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 1565,kernel_initializer = "random_uniform",activation= "relu"))
model.add(Dense(units = 3000,kernel_initializer = "random_uniform",activation= "relu"))
model.add(Dense(units = 3000,kernel_initializer = "random_uniform",activation= "relu"))
model.add(Dense(units = 1,kernel_initializer = "random_uniform",activation= "sigmoid"))
```

```python
model.compile(optimizer = "rmsprop",loss = "binary_crossentropy",metrics = ["accuracy"])
```

```python
model.fit(x_train,y_train, epochs = 5)
```

```python
pred = model.predict(x_test)
```

```python
pred = pred>0.5
```
