```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(11,22,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.xlim(11,22)
plt.ylim(11**2,22**2)
plt.plot(x,y,color = "pink")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.title("this is title")
plt.show()

```

```python
plt.xlim(11,22)
plt.ylim(11**2,22**2)
plt.scatter(x,y,color = "green")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.title("this is title")
plt.show()

```

```python

plt.plot(x,y,color = "green",label = "x vs y")
plt.plot(y,x, color = "orange", label = "y vs X")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.title("this is title")
plt.legend()
plt.show()
```

```python
plt.subplot(2,2,2)
plt.scatter(x,y,color = "green")
plt.subplot(2,2,3)
plt.plot(x,y,color = "purple")
```

```python

plt.bar(x,y,color = "green")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.title("this is title")
plt.show()

```

```python
courses = ["AI/ML","IOT","Blockchain","Python"]
students = [200,175,180,120]
```

```python
plt.pie(students,labels = courses, autopct= '%0.3f%%',colors = ["purple","pink","yellow","orange"])
plt.show()
```

```python
fig = plt.figure(figsize = (20,20))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(x,y,color = "green")
ax2.scatter(x,y,color = "red")
ax3.bar(x,y,color = "blue")
ax4.pie(students,labels = courses, autopct= '%0.3f%%',colors = ["purple","pink","yellow","orange"])
plt.show
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x="total_bill",y = "tip",data = tips)
```

```python
# the data is linearly disstributed and as the total bill inceases tip is also increasing. who ever payss hight bill they tend to give more tip
sns.lineplot(x = "size",y = "tip",data = tips)
```

```python
#size directly prop to tip # confidence interval
sns.lineplot(x = "size",y = "tip",ci =50,data = tips)
```

```python
sns.distplot(tips["tip"])
```

```python
sns.countplot(x = "smoker",data = tips)
```

```python
sns.boxplot(x ="size",y =  "tip",data = tips)
```

```python
tips.describe()
```

```python
sns.catplot(x="day",y= "tip",hue = "sex",kind = "bar",data = tips)
```
