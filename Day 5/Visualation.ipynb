```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(2, 3, 12)
y = x**2
```

```python
# plt.xlim(11,22)
# plt.ylim(11,342)

plt.plot(x, y, color = 'green')
plt.xlabel('this is x label')
plt.ylabel('this is y label')
plt.title('this is the TITLE')

#for other IDE
plt.show() #is needed

# plt.xlim(11,22)
# plt.lyim(11342)
```

```python
plt.subplot(2,2,1)
plt.scatter(x,y, color = 'green')
plt.subplot(2,2,2)
plt.plot(x,y, color = 'red')
```

```python
plt.bar(x, y, color = 'green')
plt.xlabel('this is x label')
plt.ylabel('this is y label')
plt.title('this is the TITLE')
plt.show()
```

```python
courses = ["AI and ML", "IOT", "BlockChain", "Python"]
students = [200, 175, 190, 134]
```

```python
plt.pie(students, labels = courses)
plt.show()
```

```python
plt.pie(students, labels = courses, autopct = "%0.3f%%", colors = ["purple", "red", "yellow", "orange"])
plt.show()
```

```python
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(x, y, color = "green")
ax2.scatter(x, y, color = "red")
ax3.bar(x, y, color = "purple")
ax4.pie(students, labels = courses, autopct = "%0.3f%%", colors = ["purple", "red", "yellow", "orange"])
plt.show()
```
